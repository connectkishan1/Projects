```
from keras.datasets import cifar10
((Xtrain, Ytrain), (Xtest, Ytest)) = cifar10.load_data()
```

```
print(Xtrain.shape)
print(Xtest.shape)
print(Ytrain.shape)
print(Ytest[1:5])
```

```
import matplotlib.pyplot as plt
from matplotlib import pyplot
pyplot.imshow(Xtrain[49999,:,:])
pyplot.show()
```

```
#Reshaping in 4D
Xtrain=Xtrain.reshape(Xtrain.shape[0],Xtrain.shape[1],Xtrain.shape[2],Xtrain.shape[3])
print("Xtrain_Shape",Xtrain.shape)
Xtest=Xtest.reshape(Xtest.shape[0],Xtest.shape[1],Xtest.shape[2],Xtest.shape[3])
print("Xtest_Shape",Xtest.shape)
```

```
#for precison
Xtrain=Xtrain.astype('float')
Xtest=Xtest.astype('float')

#Normailize the data 
Xtrain=Xtrain/255
Xtest=Xtest/255
```

```
#from keras.utils import to_categorical
import keras
Ytrain=keras.utils.to_categorical(Ytrain)
print(Ytrain.shape)
Ytest=keras.utils.to_categorical(Ytest)
print(Ytest.shape)
```

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D,Flatten,Dense
model = Sequential()
model.add(Conv2D(32,(5,5),padding="valid",strides=1,name="Conv1",activation="relu",input_shape=(32,32,3)))
model.add(Conv2D(64,(5,5),padding="valid",strides=1,name="Conv2",activation="relu"))
#practice model.add(Conv2D(8,(2,2),padding="valid",strides=1,name="Conv2"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(128,(5,5),padding="valid",strides=1,name="Conv3",activation="relu"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64,activation ="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(10,activation ="softmax"))
```

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D,Flatten,Dense
model = Sequential()
model.add(Conv2D(32,(8,8),padding="valid",strides=1,name="Conv1",activation="relu",input_shape=(32,32,3)))
#practice model.add(Conv2D(8,(2,2),padding="valid",strides=1,name="Conv2"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(64,(5,5),padding="valid",strides=1,name="Conv2",activation="relu"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64,activation ="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(10,activation ="softmax"))
```

```
model.summary()
```

```
model.compile(loss = keras.losses.categorical_crossentropy,optimizer = keras.optimizers.Adadelta(),metrics =['accuracy'])
model.fit(Xtrain,Ytrain,batch_size = 32,epochs =10,verbose= 1,validation_split = 0.2)
```

```
model.evaluate(Xtest,Ytest)
```
